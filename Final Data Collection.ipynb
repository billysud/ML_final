```python
# Importing libraries for different classifiers and utilities
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score

# KNN
from sklearn.neighbors import KNeighborsClassifier

# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Random Forests
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
# Using datasets from UCI Machine Learning Repository
from ucimlrepo import fetch_ucirepo 

# fetch dataset 
adult = fetch_ucirepo(id=2)
iris = fetch_ucirepo(id=53)
rice = fetch_ucirepo(id=545) 

# data (as pandas dataframes) 
x0 = adult.data.features 
y0 = adult.data.targets
x1 = iris.data.features 
y1 = iris.data.targets
x2 = rice.data.features 
y2 = rice.data.targets 
  
# variable information 
# print(adult.variables) 
# print(iris.variables)
# print(rice.variables) 
```

```python
# Logistic Regression on dataset 1
x0 = x0[['age', 'capital-gain', 'capital-loss', 'hours-per-week']]
x02_train, x02_test, y02_train, y02_test = train_test_split(x0, y0, test_size=0.8, random_state=42)
x05_train, x05_test, y05_train, y05_test = train_test_split(x0, y0, test_size=0.5, random_state=42)
x08_train, x08_test, y08_train, y08_test = train_test_split(x0, y0, test_size=0.2, random_state=42)

# Reassigning classes
y02_test['income'].replace({'<=50K': 0, '<=50K.': 0, '>50K': 1, '>50K.' : 1})
y02_train['income'].replace({'<=50K': 0, '<=50K.': 0, '>50K': 1, '>50K.' : 1})

y05_test['income'].replace({'<=50K': 0, '<=50K.': 0, '>50K': 1, '>50K.' : 1})
y05_train['income'].replace({'<=50K': 0, '<=50K.': 0, '>50K': 1, '>50K.' : 1})

y08_test['income'].replace({'<=50K': 0, '<=50K.': 0, '>50K': 1, '>50K.' : 1})
y08_train['income'].replace({'<=50K': 0, '<=50K.': 0, '>50K': 1, '>50K.' : 1})

# Logistic regression model
logreg_modelx02 = LogisticRegression()
logreg_modelx02.fit(x02_train, y02_train.values.ravel()) 
y02_pred = logreg_modelx02.predict(x02_test)

logreg_modelx05 = LogisticRegression()
logreg_modelx05.fit(x05_train, y05_train.values.ravel()) 
y05_pred = logreg_modelx05.predict(x05_test)

logreg_modelx08 = LogisticRegression()
logreg_modelx08.fit(x08_train, y08_train.values.ravel()) 
y08_pred = logreg_modelx08.predict(x08_test)

# Cross Validation
cvx02_scores = cross_val_score(logreg_modelx02, x0, y0.values.ravel(), cv=5)
cvx05_scores = cross_val_score(logreg_modelx05, x0, y0.values.ravel(), cv=5)
cvx08_scores = cross_val_score(logreg_modelx08, x0, y0.values.ravel(), cv=5)

# Evaluate testing accuracy
testingx02_accuracy = accuracy_score(y02_test, y02_pred)
testingx05_accuracy = accuracy_score(y05_test, y05_pred)
testingx08_accuracy = accuracy_score(y08_test, y08_pred)

# Evaluate training accuracy
y02_pred_train = logreg_modelx02.predict(x02_train)
y05_pred_train = logreg_modelx05.predict(x05_train)
y08_pred_train = logreg_modelx08.predict(x08_train)

trainingx02_accuracy = accuracy_score(y02_train, y02_pred_train)
trainingx05_accuracy = accuracy_score(y05_train, y05_pred_train)
trainingx08_accuracy = accuracy_score(y08_train, y08_pred_train)

print("Testing Accuracy:", testingx02_accuracy, testingx05_accuracy, testingx08_accuracy)
print("Training Accuracy:", trainingx02_accuracy, trainingx05_accuracy, trainingx08_accuracy)
print("Cross-Validation Scores:", cvx02_scores, cvx05_scores, cvx08_scores)
```

```python
print("Max Testing Accuracy:", max(testingx02_accuracy, testingx05_accuracy, testingx08_accuracy))
print("Max Training Accuracy:", max(trainingx02_accuracy, trainingx05_accuracy, trainingx08_accuracy))
print("Max mean Cross-Validation Scores:", max(cvx02_scores.mean(), cvx05_scores.mean(), cvx08_scores.mean()))
```
